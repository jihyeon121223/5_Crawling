```python
#kakaoAPI에 주소를 넣어서 위.경도 정보를 리턴하는 함수를 작성

def get_geo_code(address):
    REST_API_KEY='12c60ab21ec33500c86de172cbd60788'
    url=f'https://dapi.kakao.com/v2/local/search/keyword.json?query={address}&category_group_code=PO3'
    headers={'Authorization': f'KakaoAK {REST_API_KEY}'}
    res=requests.get(url,headers=headers)
    res=res.json()
    try:
        return res['documents'][0]['y'],res['documents'][0]['x'] #()하나,안하나 묶여서 나감
    except:
        return "결과없음"
    #return x,y #숫자형태로 리턴
```

```python
map = folium.Map([35.168786065841665, 129.06103866059343],zoom_start=20)
for index, row in df.iterrows():
    #print(index)
    #print(row['동']) 
    geodata = get_geo_code(row['동'])
    #print(geodata)
    #print('=========')
    if geodata != None and geodata != '결과없음': 
        folium.Marker(geodata,popup=row['총인구수']).add_to(map)
```

```python
import json
rfile = open('HangJeongDong_ver20220309.geojson','r',encoding='utf-8').read()
jdata = json.loads(rfile)
jsondata_busan = {'type':'FeatureCollection'}
jsondata_pick = []
for item in jdata['features']:
    #print(item)
    if item['properties']['sidonm'] == '부산광역시':
        #print(item['properties']['ELEC_EMD'])
        item['id']=item['properties']['ELEC_EMD']
        jsondata_pick.append(item)
jsondata_busan['features']=jsondata_pick
```

```python
import folium

map = folium.Map([35.168786065841665, 129.06103866059343],zoom_start=10)
#rfile = open('jeju.json','r').read()
#jdata = json.loads(rfile)
c=folium.Choropleth(jsondata_busan,
                 data=df2['총인구수'],
                 columns=[df.index,df['총인구수']],
                 key_on='feature.id',
                 fill_color='YlGn',
                 fill_opacity=0.7,
                 line_opacity=0.3,
                 legend_name='부산 인구').add_to(map)
c.geojson.add_child(
    folium.features.GeoJsonTooltip(['emd_nm'],labels=False)
)
map
```
