```python
from urllib.request import urlopen
from urllib.parse import urlencode
```

```python
api = 'https://search.naver.com/search.naver'
values = {
    'where':'nexearch',
    'sm':'top_hty',
    'fbm':1,
    'ie':'utf8',
    'query':'벗꽃'
}
```

```python
params = urlencode(values)
params
```

```python
url = api+'?'+params
url
```

```python
urlopen(url).read().decode('utf-8')
```

```python
requests.get(api,params=values).text
```
