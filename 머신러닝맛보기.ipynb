```python
from sklearn.datasets import fetch_openml #머신러닝용 . 테라스?로 접근해서 텐서플로우쓰면 편함. #데이터샘플가져오기
```

```python
mnist = fetch_openml('mnist_784')
```

```python
mnist
```

```python
mnist.data  #값이 크면클수록 밝은색(흰색)
```

```python
mnist.target
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(mnist.data,mnist.target,test_size=0.1) #0.1은 타겟값을 train 90%, test 10%로 맞춰서 분류.
```

```python
y_train.value_counts()
```

```python
y_test.value_counts()
```

```python
#이제 모델을 넣어서 학습시킬거다
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf = RandomForestClassifier() #모델생성
clf.fit(x_train,y_train) #모델에 넣어 학습시키기 
pred = clf.predict(x_test) #학습된 내용을 확인, 학습이 되었으면 응용할 수 있다
```

```python
pred #예측된 결과 보기
```

```python
#에측된 결과와 답을 맞춰보기 
#metrics는 검증용
from sklearn.metrics import accuracy_score
accuracy_score(y_test,pred)
```

```python
#시각화로 진행을 눈으로 인지

import numpy as np
import matplotlib.pyplot as plt

n_test = len(x_test)
#n_test
random_pick = np.random.randint(0,n_test,10) #10개 무작위 뽑아서 확인
#random_pick
```

```python
#뽑아낸 내용을 시각화 #미국은행에서 수치를 기계화 시키려고 만듦
figure = plt.figure(figsize=(12,5)) #가로12인치, 세로 5인치
axes=[]

for i in range(1,11): #랜덤으로 뽑은 10개를 그림그리기
    axes.append(figure.add_subplot(2,5,i)) #2행5열의 i(0~9)번째, 반복문 돌면서 안에 들어가는 subplot은 틀을 만들어 놓고
    
tmp_list=[] #비어있는 리스트 만들어주고
for i in range(10):
    tmp = x_test.iloc[random_pick[i]] #tmp에다 하나 갖고온다 #인덱스로 갖고오기 때문에 loc
    tmp = np.array(tmp)
    tmp = tmp.reshape(28,28) #reshape: 한 행씩 잘라놓은것을 연결해준다
    tmp_list.append(tmp) #데이터 10개를 다 갖고옴

print(y_test.iloc[random_pick])
for i in range(10):
    axes[i].matshow(tmp_list[i])
```
